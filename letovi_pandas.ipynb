```python
import os
import pandas as pd


cols = ["Destination", "Prices", "Departure date", "Arrival date", "Airline"]
path = os.path.join(os.getcwd(), "letovi.csv")
data = pd.read_csv(path, names=cols, skiprows=1)

data.rename_axis()
data
```

```python
print(data.shape)
#print(data.values)
print(data.loc[:, "Prices"])
```

```python
spl = list()
for price in data["Prices"]:
    #razdvoji cijenu od valute
    spl.append(price.split(" "))

prices = pd.Series(
    [float(i[0].replace(".", ""))/7.52 for i in spl],
    index = [j for j in range(len(spl))]
)
data["Prices"] = prices
data["Prices"]=data["Prices"].apply(lambda x:round(x,2))


data.insert(2, "Currency", "")
data.Currency = "EUR"


data

```

```python
#micanje dana iz datuma
data["Day"] = [i[0:4].strip() for i in data["Departure date"]]
departure = [i[4:len(i)+1] for i in data["Departure date"]]
data["Departure date"] = departure
arrival = [i[4:len(i)+1] for i in data["Arrival date"]]
data["Arrival date"] = arrival
data
```

```python
#u datetime format
str_month = {"Jan" : 1, "Feb" : 2, "Mar" : 3, "Apr" : 4, "May" : 5, "Jun" : 6, "Jul" : 7, "Aug" : 8, "Sep" : 9, "Oct" : 10, "Nov" : 11, "Dec" : 12}

for index, val in data["Departure date"].items():
    spl = val.split(" ")
    data["Departure date"][index] = spl[0] + "/" + str(str_month[spl[1]]) + "/2021"

for index, val in data["Arrival date"].items():
    spl = val.split(" ")
    data["Arrival date"][index] = spl[0] + "/" + str(str_month[spl[1]]) + "/2021"

data['Arrival date'] = pd.to_datetime(data['Arrival date'], format="%d/%m/%Y")
data['Departure date'] = pd.to_datetime(data['Departure date'], format="%d/%m/%Y")

data
```

```python
#dodavanje vremena putovanja


for index, row in data.iterrows():
    if row["Arrival date"] < row["Departure date"]:
        data["Arrival date"][index] += pd.offsets.DateOffset(years=1)

data['Travel time'] = (data['Arrival date'] - data['Departure date']).dt.days

for index, travel in data["Travel time"].items():
    data["Travel time"][index] *=24
    if travel == 0:
        data["Travel time"][index] = 1
data
```
